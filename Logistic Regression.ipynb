```python
#Linear Regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import datasets

boston=datasets.load_boston() #creating daset object

#features and labels

X=boston.data
y=boston.target

print(X.shape)
print(X)
print(y.shape)
print(y)

#creating linear regression model

l_reg=linear_model.LinearRegression()
plt.scatter(X.T[5],y)#shows that data is appropriate for linear regression

#model

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

model=l_reg.fit(X_train,y_train)
predictions=model.predict(X_test)
print('Predictions',predictions)
print('R^2',l_reg.score(X,y))#explains to what extent the variance of one variable explains the variance of the second variable.
#So, if the R2 of a model is 0.50, then approximately half of the observed variation can be explained by the model's inputs.
#R^2 is a tatistical measure of how close the data are to the fitted regression line. 

print("coedd:",l_reg.coef_)
#coefficient factor: slope of all our fit trenlines.Scale factor of multiplication of x variable(if y=mx+c,then coefficient factor is m)

print("Intercepts",l_reg.intercept_)
























```
