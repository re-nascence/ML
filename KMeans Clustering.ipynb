```python
#KMeans Clustering
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import scale
import pandas as pd
import numpy as np
# ()  when creating an object
bc=load_breast_cancer()
# print(bc,"_"*185)

X=bc.data
print(X)
print()
print()
X=scale(X)#normalize largely sparse data into a scale

print(X)

y=bc.target

print(y.shape)
print(X.shape)
plt.scatter(X.T[5],y)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

model=KMeans(n_clusters=2,random_state=0)#If we fix random state,it helps in debugging

model.fit(X_train)#We dont pass y in clustering,algorithm seperates them into different clusters by itself

predictions=model.predict(X_test)

labels=model.labels_
print("Labels: ",labels)

print("predictions: ",predictions)

print("accuracy score :",accuracy_score(y_test,predictions))

print("actual values",y_test)

#If accuracy very less,it could mean that 0's and 1's are flipped

print(pd.crosstab(y_train,labels))





























```
